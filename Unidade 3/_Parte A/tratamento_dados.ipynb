## Combinando e Mesclando Conjuntos de Dados

### Junções no DataFrame no Estilo de Bancos de Dados

```python
df1 = pd.DataFrame({'Chave': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'Coluna 1': range(7)})
df2 = pd.DataFrame({'Chave': ['a', 'b', 'd'],
                    'Coluna 2': range(3)})
print(df1)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='Chave')
```

```python
df3 = pd.DataFrame({'Chave E': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'Coluna 1': range(7)})
df4 = pd.DataFrame({'Chave D': ['a', 'b', 'd'],
                    'Coluna 2': range(3)})
pd.merge(df3, df4, left_on='Chave E', right_on='Chave D')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
df1 = pd.DataFrame({'Chave': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'Coluna 1': range(6)})
df2 = pd.DataFrame({'Chave': ['a', 'b', 'a', 'b', 'd'],
                    'Coluna 2': range(5)})
print(df1)
print(df2)

pd.merge(df1, df2, on='Chave', how='left')
```

```python
pd.merge(df1, df2, how='inner')
```

```python
left = pd.DataFrame({'Chave1': ['foo', 'foo', 'bar'],
                     'Chave2': ['one', 'two', 'one'],
                     'ValorE': [1, 2, 3]})
right = pd.DataFrame({'Chave1': ['foo', 'foo', 'bar', 'bar'],
                      'Chave2': ['one', 'one', 'one', 'two'],
                      'ValorR': [4, 5, 6, 7]})
print(left)
print(right)
pd.merge(left, right, on=['Chave1', 'Chave2'], how='outer')
```

```python
#pd.merge(left, right, on='Chave1')
pd.merge(left, right, on='Chave1', suffixes=('_left', '_right'))
```

### Merge com Base no Índice

```python
left1 = pd.DataFrame({'Chave': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'Valor': range(6)})
right1 = pd.DataFrame({'Grupo_val': [3.5, 7]}, index=['a', 'b'])
print(left1)
print(right1)
pd.merge(left1, right1, left_on='Chave', right_index=True)
```

```python
pd.merge(left1, right1, left_on='Chave', right_index=True, how='outer')
```

```python
lefth = pd.DataFrame({'Chave1': ['SC', 'SC', 'SC',
                               'RS', 'RS'],
                      'Chave2': [2000, 2001, 2002, 2001, 2002],
                      'Dados': np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['RS', 'RS', 'SC', 'SC',
                              'SC', 'SC'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['Evento1', 'Evento2'])
print(lefth)
print(righth)
```

```python
res=pd.merge(lefth, righth, left_on=['Chave1', 'Chave2'], right_index=True)
print(res)
pd.merge(lefth, righth, left_on=['Chave1', 'Chave2'], right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=['a', 'c', 'e'],
                     columns=['SC', 'RS'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=['b', 'c', 'd', 'e'],
                      columns=['PR', 'SP'])
print(left2)
print(right2)
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

```python
print(left1)
print(right1)
left1.join(right1, on='Chave')
```

### Concatenando ao Longo de um Eixo

```python
arr = np.arange(12).reshape((3, 4))
print(arr)
np.concatenate([arr, arr], axis=1)
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
pd.concat([s1, s2, s3])
```

### Combinando Dados com Sobreposição

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
              index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
              index=['f', 'e', 'd', 'c', 'b', 'a'])
#b[-1] = np.nan
print(a)
print(b)
np.where(pd.isnull(a), b, a)
```

```python
a.combine_first(b)
```

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 16, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})
print(df1)
print(df2)
df1.combine_first(df2)
```
